```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file containing actual and predicted J-D values
data = pd.read_csv('actual_vs_predicted_jd_co2.csv')

# Extract the actual and predicted J-D values
actual_jd = data['Actual'].values
predicted_jd = data['Predicted'].values

# Create a scatter plot of Predicted J-D vs Actual J-D
plt.figure(figsize=(8, 6))
plt.scatter(actual_jd, predicted_jd, color='red', label='Predicted vs Actual')

# Add labels and a title
plt.xlabel('Actual J-D')
plt.ylabel('Predicted J-D')
plt.title('Predicted vs Actual J-D Values')

# Optionally add a line representing perfect prediction
plt.plot([min(actual_jd), max(actual_jd)], [min(actual_jd), max(actual_jd)], color='blue', linestyle='--', label='Perfect Prediction')

# Display the legend and grid
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
