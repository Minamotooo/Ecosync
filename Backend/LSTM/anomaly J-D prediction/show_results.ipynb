```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load data from CSV
data = pd.read_csv('actual_vs_predicted_jd.csv')

# Assuming the CSV has two columns: 'Actual' and 'Predicted'
actual_jd = data['Actual J-D'].values
predicted_jd = data['Predicted J-D'].values

# Step 2: Create scatter plot of Actual vs Predicted J-D
plt.scatter(actual_jd, predicted_jd, label="Predictions", color='red')

# Step 3: Plot ideal fit line for reference
plt.plot([min(actual_jd), max(actual_jd)], [min(actual_jd), max(actual_jd)], 
         color='blue', linestyle='--', label="Ideal Fit")

# Step 4: Set labels for axes
plt.xlabel('Actual J-D')
plt.ylabel('Predicted J-D')

ticks = np.arange(0.8, 1.3, 0.05)
plt.xticks(ticks)  # Set x-axis ticks
plt.yticks(ticks)  # Set y-axis ticks

# Step 6: Add legend and title
plt.legend()
plt.title('Actual vs Predicted J-D')

# Step 7: Show the plot
plt.show()
```
