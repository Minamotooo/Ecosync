```python
# **Number of Landslides per Year**
landslides_per_year = new_df.groupby('year').size().reset_index(name='count')
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=landslides_per_year, x='year', y='count', marker='o')
plt.title('Number of Landslides per Year')
plt.xlabel('Year')
plt.ylabel('Number of Landslides')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# **Distribution of Landslide Triggers**
plt.figure(figsize=(12, 8))
sns.countplot(y='landslide_trigger', data=new_df, order=new_df['landslide_trigger'].value_counts().index)
plt.title('Distribution of Landslide Triggers')
plt.xlabel('Count')
plt.ylabel('Landslide Trigger')
plt.show()
```

```python
# **Fatalities Over Time**
fatalities_per_year = new_df.groupby('year')['fatality_count'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=fatalities_per_year, x='year', y='fatality_count', marker='o', color='red')
plt.title('Total Fatalities from Landslides per Year')
plt.xlabel('Year')
plt.ylabel('Total Fatalities')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python

# **Geographical Distribution of Landslides**

# Map 'country_name' to ISO alpha-3 codes for mapping
def get_country_code(name):
    try:
        return pycountry.countries.lookup(name).alpha_3
    except LookupError:
        return None
```

```python
# **Analyzing Landslides Related to Climate Change Factors**

# Standardize 'landslide_trigger' entries to lowercase
new_df['landslide_trigger'] = new_df['landslide_trigger'].str.lower()

# Define triggers related to climate change
rain_triggers = ['rainfall', 'heavy rainfall', 'monsoon', 'rain', 'extreme rainfall']

# Filter data for rainfall-related landslides
rainfall_landslides = new_df[new_df['landslide_trigger'].isin(rain_triggers)]
rainfall_per_year = rainfall_landslides.groupby('year').size().reset_index(name='count')
```

```python
 plt.figure(figsize=(12, 6))
sns.lineplot(data=rainfall_per_year, x='year', y='count', marker='o', color='blue')
plt.title('Landslides Triggered by Rainfall per Year')
plt.xlabel('Year')
plt.ylabel('Number of Landslides')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# **Fatalities by Top Landslide Triggers Over Time**
# Identify top 5 triggers
top_triggers = new_df['landslide_trigger'].value_counts().nlargest(5).index
fatalities_by_trigger = new_df[new_df['landslide_trigger'].isin(top_triggers)]
fatalities_by_trigger = fatalities_by_trigger.groupby(['year', 'landslide_trigger'])['fatality_count'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(
    data=fatalities_by_trigger,
    x='year',
    y='fatality_count',
    hue='landslide_trigger',
    marker='o'
)
plt.title('Fatalities by Top Landslide Triggers Over Time')
plt.xlabel('Year')
plt.ylabel('Total Fatalities')
plt.legend(title='Landslide Trigger', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# **Conclusion**
# The analysis provides insights into how climate change factors like increased rainfall influence landslide occurrences and fatalities over time.
```

```python
# Filter the data for years from 2000 onwards
fatalities_by_trigger_2000 = fatalities_by_trigger[fatalities_by_trigger['year'] >= 2000]

# Now plot using the filtered data
plt.figure(figsize=(12, 8))
sns.lineplot(
    data=fatalities_by_trigger_2000,
    x='year',
    y='fatality_count',
    hue='landslide_trigger',
    marker='o'
)
plt.title('Fatalities by Top Landslide Triggers Over Time (2000 onwards)')
plt.xlabel('Year')
plt.ylabel('Total Fatalities')
plt.legend(title='Landslide Trigger', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
